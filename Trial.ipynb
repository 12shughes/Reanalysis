```python
def calculate_pfull(psurf, ak, bk):
    r"""Calculates full pressures using surface pressures and sigma coordinates

    psurf  : array-like
            Surface pressures
    siglev : array-like
            Sigma-levels
    """

    ph = psurf * bk + ak
    ps = ph.shift(phalf = -1)
    pf = (ps - ph)/np.log(ps/ph)

    return pf
```
